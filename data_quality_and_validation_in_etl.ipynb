**Question 1 : Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?**
- Data Quality means making sure the data is correct, complete, consistent, and reliable before using it for reports or analysis.

It is more than just cleaning because:

Cleaning only fixes errors (like NULL values or wrong spelling).

Data quality also checks accuracy, duplicates, format, consistency, and business rules.

It ensures the data is trustworthy for decision-making.

**Question 5 : Why should data validation be performed during transformation rather than after loading?**
- if we load wrong data first, it spreads errors.

Cleaning later is difficult and risky.

Validation during transformation saves time.

It prevents bad data from entering final database.

So better to stop wrong data before loading.
